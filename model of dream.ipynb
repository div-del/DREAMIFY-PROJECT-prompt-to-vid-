```python
from diffusers import StableDiffusionPipeline
import torch
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", safety_checker=None)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
out = pipe("a luxurious bathroom in a rich bunglow with a couple", guidance_scale=7, num_inference_steps=30)
out.images[0].save("sample.png")
print("Saved sample.png")
```
